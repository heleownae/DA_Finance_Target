## 2-1.군집별 한 번에 보기

```python
unique_group = sorted(son_cos_py['코사인군집'].unique())
unique_group
```

```python
palette = sns.color_palette('Pastel1')
sns.set(style="whitegrid", font_scale=1)
plt.rc('font', family='NanumGothic')
```

### 2-1-1. 성별

```python
# 서브플롯 준비
fig, axes = plt.subplots(nrows=1, ncols=len(unique_group), figsize=(12, 3), sharey=True)

# 각 유니크한 're_group' 값에 대해 서브플롯 생성
for ax, group in zip(axes, unique_group):
    # 현재 cycle 값에 해당하는 데이터 필터링
    subset = son_cos_py[son_cos_py['코사인군집'] == group]

    # countplot 그리기
    sns.countplot(x='GENDER', data=subset, ax=ax, palette=palette)
    ax.set_title(f'군집 {group}')
    ax.set_xlabel('성별')
    ax.set_ylabel('Frequency')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
total_counts = son_cos_py['코사인군집'].value_counts().rename('total')

# 건수 계산
counts = son_cos_py.groupby(['코사인군집', 'GENDER']).size().unstack(fill_value=0)

# 군집별 전체 건수 대비 건수 비율 계산
merged = counts.merge(total_counts, left_index=True, right_index=True)
for gender in counts.columns:
    merged[gender] = merged[gender] / merged['total']

# 비율 시각화, stacked bar chart
fig, ax = plt.subplots(figsize=(8, 4))
merged.drop(columns='total').plot(kind='bar', stacked=True, ax=ax)
ax.set_title('군집별 성별 비율')
ax.set_xlabel('코사인 군집')
ax.set_ylabel('비율')
ax.legend(title='성별')
plt.tight_layout()
plt.show()
```

```python
total_counts = son_cos_py['코사인군집'].value_counts().rename('total')

# 건수 계산
counts = son_cos_py.groupby(['코사인군집', 'GENDER']).size().unstack()   # fill_value=0

# 군집별 전체 건수 대비 건수 비율 계산
merged = counts.merge(total_counts, left_index=True, right_index=True)
for feature in counts.columns:
    merged[feature] = merged[feature] / merged['total']

# 비율 계산된 데이터프레임 'merged'를 사용
merged = merged.drop(columns='total').stack().reset_index()
merged.columns = ['코사인군집', 'GENDER', '비율']

# seaborn의 catplot을 사용하여 stacked bar chart 비슷한 효과 구현
g = sns.catplot(x='코사인군집', y='비율', hue='GENDER', data=merged, kind='bar', height=4, aspect=2, palette=palette)

# catplot으로 자동 생성된 범례 지우기
g._legend.remove()

# 세부 설정
plt.title('군집별 성별 비율')
plt.xlabel('코사인 군집')
plt.ylabel('비율')
plt.legend(title='성별', bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
