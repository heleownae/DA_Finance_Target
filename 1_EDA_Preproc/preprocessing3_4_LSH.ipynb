```python
cont_pre3['INS_GBN_CD'].info()
```

```python
#범주 축소
def GBN_CN_reduce(x):
  match x:
    case 1|2:
        return 1
    case 3|5:
        return 2
    case 4:
        return 3
    case 7|16:
        return 4
    case 13:
        return 5
    case 14:
        return 6
    case 9|10|11:
        return 7
    case 6|8|12|15|99:
        return 8
    case 18:
        return 9

cont_pre3['reduce_GBN_CD'] = cont_pre3['INS_GBN_CD'].apply(GBN_CN_reduce)
cont_pre3
```
