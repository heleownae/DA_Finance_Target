# 2.보험금 지급 내용까지 결합

## 2-2. 보험금 지급이 없는 데이터는 모두 0으로 진행
  * 결합된 테이블 : merge_all3

```python
merge_all2 = pd.merge(merge_p_cont, pmnt_pol, on=['JOIN_SN', 'POL_SN'], how='left').sort_values('POL_SN')
merge_all2
```

```python
merge_all2.isnull().sum()
```

```python
merge_all3 = merge_all2.drop(columns=['INS_GBN_CD'])
merge_all3
```

```python
merge_all3.fillna(0, inplace=True)
merge_all3
```

```python
to_int = merge_all3.columns[-10:]
to_int
```

```python
merge_all3[to_int] = merge_all3[to_int].astype(int)
merge_all3
```
