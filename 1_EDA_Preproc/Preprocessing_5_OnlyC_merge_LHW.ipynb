```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn 

plt.rcParams['font.family'] ='D2Coding'
pd.set_option('display.max_columns', None)
```

```python
customer = pd.read_csv('customer.csv') # 차주정보
rlps = pd.read_csv('rlps_onlyC.csv') # 보험계약관계자정보
```

```python
onlyc_merged = pd.merge(rlps, customer, on=['JOIN_SN'], how='left', indicator=True)
onlyc_merged.head(20)
```

```python
onlyc_merged.info()
```

```python
# 시각화
# JOIN_SN	POL_SN	CT_RLPS_GBN_CD	CT_IN_RLTN_CD	GENDER	AGE	AGE_GROUP	_merge

vis_list = ['CT_RLPS_GBN_CD', 'CT_IN_RLTN_CD', 'GENDER', 'AGE_GROUP']

def ctgr_value_counts(iumn_name):
    for uniq in list(onlyc_merged[iumn_name].unique()):
        cond_unique = (onlyc_merged[iumn_name]==uniq)
        print(uniq)
        print(onlyc_merged.loc[cond_unique]['AGE_GROUP'].value_counts())
        print()

for i in vis_list:
    plt.figure(figsize=(10,6))
    sns.countplot(data=onlyc_merged, x=i, hue='AGE_GROUP', hue_order=(0, 1, 2, 3, 4, 5, 6, 7, 8))     
    plt.legend(loc='best')
    plt.show()
    ctgr_value_counts(iumn_name=i)
```

```python
# 상관관계
# CT_IN_RLTN_CD	GENDER	AGE	AGE_GROUP

corr_df = onlyc_merged[['CT_IN_RLTN_CD', 'GENDER', 'AGE_GROUP']]

plt.rcParams["figure.figsize"] = (16, 12)
sns.heatmap(corr_df.corr(), annot = True, annot_kws={"size":8})
plt.show()
```
