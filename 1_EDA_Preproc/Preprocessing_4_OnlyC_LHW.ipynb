```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn 

plt.rcParams['font.family'] ='D2Coding'
pd.set_option('display.max_columns', None)
```

```python
rlps = pd.read_csv('SDB2_I_CT_RLPS.csv') # 보험계약관계자정보
company = pd.read_csv('SDB2_A_DESC.csv') #기업정보
```

```python
rlps = rlps[['JOIN_SN', 'JOIN_SN_TYP', 'POL_SN', 'CT_RLPS_GBN_CD', 'CT_IN_RLTN_CD', 'IS_INDIV']]
rlps.head()
```

```python
rlps = rlps[rlps['IS_INDIV'] == 1]
rlps['IS_INDIV'].value_counts()
```

```python
rlps = rlps[['JOIN_SN', 'JOIN_SN_TYP', 'POL_SN', 'CT_RLPS_GBN_CD', 'CT_IN_RLTN_CD']]
rlps.head()
```

```python
rlps = rlps[['JOIN_SN', 'POL_SN', 'CT_RLPS_GBN_CD', 'CT_IN_RLTN_CD']]
rlps.head()
```

```python
personal = company[company['JOIN_SN_TYP'] == 1]
personal.head()
```

```python
rlps = rlps[~rlps['JOIN_SN'].isin(personal['JOIN_SN'])]
display(rlps.info())
display(rlps.head())
```

```python
rlps_onlyC = rlps[rlps['CT_RLPS_GBN_CD'] == 'C']
rlps_onlyC.head(20)
```

```python
rlps_onlyC.info()
```

```python
# JOIN_SN, POL_SN 둘 다 같은 경우만 일단 추출

rlps_dup = rlps_onlyC[rlps_onlyC.duplicated(subset=['JOIN_SN', 'POL_SN'], keep=False)]
rlps_dup = rlps_dup.sort_values(by=['JOIN_SN', 'POL_SN'])
display(rlps_dup.info())
display(rlps_dup.head(20))
```

```python
rlps_duptouniq = rlps_dup.drop_duplicates()
rlps_duptouniq.info()
```

```python
# rlps_onlyC - rlps_dup
rlps_onlyC2 = rlps_onlyC[~rlps_onlyC.isin(rlps_dup).all(1)]

# rlps_onlyC2 + rlps_duptouniq
result = pd.concat([rlps_onlyC2, rlps_duptouniq], ignore_index=True)

result.head(20)
```
